```python
#Assignment on orders.txt data set

#1- create a new column in the orders dataframe with name profit_flag. Set its value as 1 if profit is greater than 0 else set 0.
def flag(val):
    if val>0:
        return 1
    else:
        0
```

```python
df['profit_falg']=df['profit'].apply(flag)
df
```

```python
#in the orders data frame rename column names to upper case ..eg order_id should be ORDER_ID
df.columns=df.columns.str.upper()
df
```

```python
#3- again change the column names back to lower case
df.columns=df.columns.str.lower()
df
```

```python
#sort the dataframe column by profit in descending order
df_new = df.sort_values(by='profit', ascending=False)
df_new
```

```python
#sort the dataframe column by sales in ascending order , in case of a tie it should be sorted by profit in descending order
df_sales_profit = df.sort_values(by = ['sales','profit'], ascending=[True,False])
df_sales_profit
```

```python
# create a column as profit_category . sets it value as below:

def category(val):
    if val<0:
        return 'loss'
    elif val>=0 and val<=100:
        return 'low profit'
    elif val>100 and val<=500:
        return 'medium profit'
    else:
        return 'high profit'
```

```python
df['profit_category'] = df['profit'].apply(category)
df
```

```python
fc = (df['profit_category']=='loss')
df[fc].sort_values(by = 'profit')
```

```python
# create a new column sales_profit_ratio and set its value as (profit/sales)*100
df['sales_profit_ratio'] = (df['profit']/df['sales'])*100
df
```

```python
# find all the orders where customer name starts with 'D' and ends with 'e'
fc = df['customer_name'].str.startswith('D') & df['customer_name'].str.endswith('e')
df[fc]
```

```python
#delete column sales_profit_ratio from the dataframe
df.drop(columns = 'sales_profit_ratio',inplace = True)
```

```python
#delete all the rows where profit is negative
fc = (df['profit']<0)
df.drop(df[fc].index,inplace = True)
```
